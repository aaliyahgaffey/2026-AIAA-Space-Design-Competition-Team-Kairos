Prior to running code, download data using this link: https://www.kaggle.com/datasets/fhnw-i4ds/sdobenchmark


After downloading the data, place it in the same directory as this Jupyter notebook.

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
from datetime import datetime

# configuration
DATA_ROOT = "SDOBenchmark_full"
BATCH_SIZE = 8          
NUM_EPOCHS = 30
LEARNING_RATE = 1e-4    
HOUR_BATCH = 1          
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

goes_classes = ['quiet', 'A', 'B', 'C', 'M', 'X']

# utilities
def flux_to_class(f: float, only_main=False):
    if f <= 1e-9: return 'quiet'
    decade = int(np.floor(np.log10(f)))
    idx = max(0, min(decade + 9, len(goes_classes) - 1))
    main_class = goes_classes[idx]
    if main_class == 'quiet' or only_main: return main_class
    sub_class = str(round(10 ** -decade * f))
    return main_class + sub_class

# --- DATASET ---
class SolarFlareDataset(Dataset):
    def __init__(self, split='training', hour_batch=1):
        self.root = os.path.join(DATA_ROOT, split)
        self.df = pd.read_csv(os.path.join(self.root, "meta_data.csv"))
        self.hour_batch = hour_batch
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
        
    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        try:
            row = self.df.iloc[idx]
            sample_id = row['id']
            ar_num = sample_id.split('_')[0]
            folder_name = "_".join(sample_id.split('_')[1:])
            path = os.path.join(self.root, ar_num, folder_name)
            
            if not os.path.exists(path):
                return self.__getitem__(np.random.randint(0, len(self.df)))

            all_imgs = sorted([f for f in os.listdir(path) if f.endswith(('.jpg', '.png'))])
            start_idx = self.hour_batch * 10
            batch_files = all_imgs[start_idx : start_idx + 10]
            
            if len(batch_files) < 10:
                return self.__getitem__(np.random.randint(0, len(self.df)))
            
            images = [self.transform(Image.open(os.path.join(path, f)).convert('RGB')) for f in batch_files]
            
            # target is now a class index (0-5)
            class_label = flux_to_class(row['peak_flux'], only_main=True)
            target_idx = goes_classes.index(class_label)

            return torch.stack(images), torch.tensor(target_idx, dtype=torch.long)
        except Exception:
            return self.__getitem__(np.random.randint(0, len(self.df)))

# ResNet50 model
class FlarePredictor(nn.Module):
    def __init__(self):
        super(FlarePredictor, self).__init__()
        base_model = models.resnet50(weights='DEFAULT') 
        self.feature_extractor = nn.Sequential(*list(base_model.children())[:-1]) 
        
        self.temporal_aggregator = nn.Sequential(
            nn.Linear(2048 * 10, 1024),
            nn.ReLU(),
            nn.Dropout(0.3)
        )
        self.regressor = nn.Linear(1024, 6) # classification output

    def forward(self, x):
        b, t, c, h, w = x.shape
        x = x.view(b * t, c, h, w)
        features = self.feature_extractor(x).view(b, t, 2048) 
        bottleneck = self.temporal_aggregator(features.view(b, -1))
        return self.regressor(bottleneck)

    def save_model(self, model_dir="models"):
        """Saves model state_dict with a YYYYMMDDHHMMSS timestamp."""
        if not os.path.exists(model_dir):
            os.makedirs(model_dir)
            
        timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
        filename = f"flare_predictor_{timestamp}.pt"
        save_path = os.path.join(model_dir, filename)
        
        torch.save(self.state_dict(), save_path)
        print(f"Model saved to: {save_path}")
        return save_path

    def load_model(self, file_path):
        """Loads a model state_dict from a specific file path."""
        if os.path.exists(file_path):
            self.load_state_dict(torch.load(file_path))
            self.eval() # set to evaluation mode after loading
            print(f"Model loaded successfully from: {file_path}")
        else:
            print(f"Error: No file found at {file_path}")

# balanced resampling
def get_balanced_df(df, samples_per_class=600):
    df['class_label'] = df['peak_flux'].apply(flux_to_class, only_main=True)
    balanced_list = []
    for cls in goes_classes:
        subset = df[df['class_label'] == cls]
        if len(subset) <= samples_per_class:
            balanced_list.append(subset)
        else:
            balanced_list.append(subset.sample(n=samples_per_class, random_state=42))
    return pd.concat(balanced_list).sample(frac=1).reset_index(drop=True)

if __name__ == '__main__':
    train_ds = SolarFlareDataset(split='training', hour_batch=HOUR_BATCH)
    test_ds = SolarFlareDataset(split='test', hour_batch=HOUR_BATCH)

    # balanced resampling
    print("Applying balanced resampling...")
    full_train_meta = pd.read_csv(os.path.join(DATA_ROOT, "training", "meta_data.csv"))
    train_ds.df = get_balanced_df(full_train_meta, samples_per_class=1000)

    train_samples = len(train_ds)
    test_samples = len(test_ds)

    print(f"--- Dataset Statistics ---")
    print(f"Training Samples: {train_samples}")
    print(f"Testing Samples:  {test_samples}")
    print(f"Total Images (Train): {train_samples * 10}") # each sample has 10 images

    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=0)
    test_loader = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=0)

    model = FlarePredictor().to(DEVICE)

    # weighted loss & optimization
    weights = torch.tensor([1.0, 1.0, 1.0, 1.0, 4.0, 10.0]).to(DEVICE)
    criterion = nn.CrossEntropyLoss(weight=weights)
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=2, factor=0.5)

    # training Loop
    history = {'train_loss': []}
    for epoch in range(NUM_EPOCHS):
        model.train()
        running_loss = 0.0
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS}")
        
        for imgs, targets in pbar:
            imgs, targets = imgs.to(DEVICE), targets.to(DEVICE)
            optimizer.zero_grad()
            outputs = model(imgs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            pbar.set_postfix(loss=f"{loss.item():.4f}")
        
        # calculate epoch loss and update scheduler
        epoch_loss = running_loss / len(train_loader)
        history['train_loss'].append(epoch_loss)
        scheduler.step(epoch_loss) # <--- Correct placement
        
        print(f"Epoch {epoch+1} Complete. Avg Loss: {epoch_loss:.4f}")

    model.save_model(model_dir="solar_checkpoints")
    # evaluation
    model.eval()
    all_preds, all_targets = [], []
    with torch.no_grad():
        for imgs, targets in tqdm(test_loader, desc="Testing"):
            outputs = model(imgs.to(DEVICE))
            _, predicted_indices = torch.max(outputs, 1)
            
            all_preds.extend(predicted_indices.cpu().numpy())
            all_targets.extend(targets.numpy()) 

    # visualization
    y_true_labels = [goes_classes[idx] for idx in all_targets]
    y_pred_labels = [goes_classes[idx] for idx in all_preds]

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1); plt.plot(history['train_loss']); plt.title("Weighted Training Loss")
    plt.subplot(1, 2, 2)
    cm = confusion_matrix(y_true_labels, y_pred_labels, labels=goes_classes)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=goes_classes)
    disp.plot(cmap='Blues', ax=plt.gca())
    plt.title("Weighted Flare Class Confusion Matrix"); plt.show()
    
    print("\nClassification Report:\n", classification_report(y_true_labels, y_pred_labels, labels=goes_classes, zero_division=0))
```
